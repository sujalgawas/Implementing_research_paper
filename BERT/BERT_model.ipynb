```python
from datasets import load_dataset

dataset = load_dataset("wikimedia/wikipedia", "20231101.en")

dataset = dataset["train"][:5000]
```

```python
import pandas as pd

df = pd.DataFrame(dataset)
df.head()
```

```python
with open('data.txt', 'w') as f:
    for x in df['text']:
        f.write(x + "\n")
```

```python
from tokenizers import Tokenizer
from tokenizers.models import WordPiece
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.trainers import WordPieceTrainer

VOCAB_SIZE = 8000

tokenizer = Tokenizer(WordPiece(unk_token="[UNK]"))
tokenizer.pre_tokenizer = Whitespace()

trainer = WordPieceTrainer(
    vocab_size=VOCAB_SIZE,
    special_tokens=[
        "[UNK]",
        "[CLS]",
        "[SEP]",
        "[PAD]",
        "[MASK]",
    ]
)

tokenizer.train(["data.txt"], trainer)

MASK_TOKEN_ID = tokenizer.token_to_id("[MASK]")
PAD_TOKEN_ID  = tokenizer.token_to_id("[PAD]")
CLS_TOKEN_ID  = tokenizer.token_to_id("[CLS]")
SEP_TOKEN_ID  = tokenizer.token_to_id("[SEP]")
```

```python
from torch.utils.data import Dataset
import torch

class BERTDataset(Dataset):
    def __init__(self, df, max_token=128, mlm_prob=0.15):
        self.df = df
        self.max_token = max_token
        self.mlm_prob = mlm_prob

    def __len__(self):
        return len(self.df)

    def __pad(self, ids):

        ids = ids[:self.max_token]
        ids = ids + [PAD_TOKEN_ID] * (self.max_token - len(ids))
        return ids

    def __getitem__(self, idx):
        txt = self.df["text"][idx]

        encoding = tokenizer.encode(txt)
        ids = encoding.ids

        ids = [CLS_TOKEN_ID] + ids + [SEP_TOKEN_ID]
        ids = self.__pad(ids)

        input_ids = torch.tensor(ids, dtype=torch.long)

        labels = input_ids.clone()

        # never mask CLS / SEP / PAD — only real content tokens
        special = (input_ids == CLS_TOKEN_ID) | (input_ids == SEP_TOKEN_ID) | (input_ids == PAD_TOKEN_ID)

        prob = torch.full(input_ids.shape, self.mlm_prob)
        prob.masked_fill_(special, 0.0)
        masked_indices = torch.bernoulli(prob).bool()

        labels[~masked_indices] = -100

        # of the masked positions: 80% -> [MASK], 10% -> random token, 10% -> left unchanged
        replace_with_mask = torch.bernoulli(torch.full(input_ids.shape, 0.8)).bool() & masked_indices
        input_ids[replace_with_mask] = MASK_TOKEN_ID

        replace_with_random = torch.bernoulli(torch.full(input_ids.shape, 0.5)).bool() & masked_indices & ~replace_with_mask
        random_tokens = torch.randint(VOCAB_SIZE, input_ids.shape, dtype=torch.long)
        input_ids[replace_with_random] = random_tokens[replace_with_random]

        segment_ids = torch.zeros_like(input_ids)

        # no padding mask existed before. True = real token, False = [PAD].
        attention_mask = (input_ids != PAD_TOKEN_ID)

        return input_ids, segment_ids, attention_mask, labels
```

```python
import torch.nn as nn
import torch

class BertEmbeddings(nn.Module):
    def __init__(self, vocab_size, hidden_dim, max_seq_len, pad_token_id=0):
        super().__init__()
        self.token_embedding    = nn.Embedding(vocab_size, hidden_dim, padding_idx=pad_token_id)
        self.segment_embedding  = nn.Embedding(2, hidden_dim)
        self.position_embedding = nn.Embedding(max_seq_len, hidden_dim)

    def forward(self, input_ids, segment_ids):
        seq_len = input_ids.size(1)
        position_ids = torch.arange(seq_len, device=input_ids.device).unsqueeze(0)

        tok_emb = self.token_embedding(input_ids)
        seg_emb = self.segment_embedding(segment_ids)
        pos_emb = self.position_embedding(position_ids)

        return tok_emb + seg_emb + pos_emb
```

```python
class FeedForward(nn.Module):
    def __init__(self, input_layer, hidden_layer, output_layer):
        super().__init__()
        self.Linear1 = nn.Linear(in_features=input_layer, out_features=hidden_layer)
        self.Linear2 = nn.Linear(in_features=hidden_layer, out_features=output_layer)
        self.Gelu = nn.GELU()

    def forward(self, x):
        return self.Linear2(self.Gelu(self.Linear1(x)))
```

```python
def get_padding_mask(attention_mask):
    return ~attention_mask
```

```python
class Transformer(nn.Module):
    def __init__(self, d_model, num_head):
        super().__init__()
        self.feed_forward = FeedForward(input_layer=d_model, hidden_layer=4096, output_layer=d_model)

        self.self_attn = nn.MultiheadAttention(embed_dim=d_model, num_heads=num_head, batch_first=True)

        self.layer_norm1 = nn.LayerNorm(normalized_shape=d_model)
        self.layer_norm2 = nn.LayerNorm(normalized_shape=d_model)

    def forward(self, x, key_padding_mask):
        attn_out, _ = self.self_attn(x, x, x, key_padding_mask=key_padding_mask)

        x = self.layer_norm1(x + attn_out)

        ff_out = self.feed_forward(x)
        x = self.layer_norm2(x + ff_out)

        return x
```

```python
class BERT(nn.Module):
    def __init__(self, vocab_size, d_model, num_head):
        super().__init__()
        self.embeddings = BertEmbeddings(vocab_size=vocab_size, hidden_dim=d_model, max_seq_len=512)
        self.transformer = nn.ModuleList([
            Transformer(d_model=d_model, num_head=num_head)
            for _ in range(12)
        ])

        self.mlm_head = nn.Linear(d_model, vocab_size)

    def forward(self, input_ids, segment_ids, key_padding_mask):
        x = self.embeddings(input_ids, segment_ids)

        for block in self.transformer:
            x = block(x, key_padding_mask)

        logits = self.mlm_head(x)
        return logits
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
from torch.utils.data import DataLoader, random_split

wiki_corpus_dataset = BERTDataset(df)

length = len(wiki_corpus_dataset)
train_len = int(0.8 * length)
val_len = length - train_len

train_dataset, val_dataset = random_split(wiki_corpus_dataset, [train_len, val_len])
train_dataloader = DataLoader(train_dataset, batch_size=128, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=128, shuffle=False)

model = BERT(vocab_size=VOCAB_SIZE, d_model=768, num_head=4).to(device)
```

```python
import torch.optim as optim

optimizer = optim.AdamW(model.parameters(), lr=5e-5, weight_decay=0.01)
loss_fn = nn.CrossEntropyLoss()

epochs = 5

for epoch in range(1, epochs + 1):
    model.train()
    train_loss = 0.0
    for input_ids, segment_ids, attention_mask, labels in train_dataloader:
        input_ids = input_ids.to(device)
        segment_ids = segment_ids.to(device)
        labels = labels.to(device)
        key_padding_mask = get_padding_mask(attention_mask).to(device)

        logits = model(input_ids, segment_ids, key_padding_mask)

        loss = loss_fn(logits.view(-1, logits.size(-1)), labels.view(-1))

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
    print(f"Epoch {epoch}, train loss: {train_loss/len(train_dataloader):.4f}")

    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for input_ids, segment_ids, attention_mask, labels in val_dataloader:
            input_ids = input_ids.to(device)
            segment_ids = segment_ids.to(device)
            labels = labels.to(device)
            key_padding_mask = get_padding_mask(attention_mask).to(device)

            logits = model(input_ids, segment_ids, key_padding_mask)
            loss = loss_fn(logits.view(-1, logits.size(-1)), labels.view(-1))
            val_loss += loss.item()
    print(f"Epoch {epoch}, val loss: {val_loss/len(val_dataloader):.4f}")
```

```python
@torch.inference_mode()
def fill_mask(txt, model):
    model.eval()
    encoding = tokenizer.encode(txt)
    ids = [CLS_TOKEN_ID] + encoding.ids + [SEP_TOKEN_ID]

    input_ids = torch.tensor([ids], dtype=torch.long, device=device)
    segment_ids = torch.zeros_like(input_ids)
    key_padding_mask = torch.zeros_like(input_ids, dtype=torch.bool)

    logits = model(input_ids, segment_ids, key_padding_mask)

    predicted_ids = input_ids.clone()
    mask_positions = (input_ids[0] == MASK_TOKEN_ID).nonzero(as_tuple=True)[0]
    for pos in mask_positions:
        predicted_ids[0, pos] = logits[0, pos].argmax(dim=-1)

    return tokenizer.decode(predicted_ids[0].tolist())

print(fill_mask("the capital of france is [MASK] .", model))
```
